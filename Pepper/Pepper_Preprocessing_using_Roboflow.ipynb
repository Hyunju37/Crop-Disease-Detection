```python
crop_names = {
    3: "파프리카",
    2: "토마토",
    5: "고추",
    4: "오이",
    1: "딸기"
}
```

```python
unique_names = set()
def extract_class_names(annotation):
    disease_code = annotation["annotations"]["disease"]
    crop_code = crop_names.get(annotation["annotations"]["crop"], "Unknown")
    #risk_code = str(annotation["annotations"]["risk"])
    #combined_class = f"{disease_code}_{crop_code}_r{risk_code}"
    combined_class = f"{disease_code}_{crop_code}"
    unique_names.add(combined_class)
```

```python
def return_class_name(annotation):
    disease_code = annotation["annotations"]["disease"]
    crop_code = crop_names.get(annotation["annotations"]["crop"], "Unknown")
    combined_class = f"{disease_code}_{crop_code}"
    return combined_class
```

```python
IBM = {
   "version": "1.0",
    "type": "localization",
    "labels": categories_ilst,
    "annotations":{}
}
id = 0
for lbl_file in all_annots[3]:
  if lbl_file.endswith(".json"):
    if lbl_file.replace('.json','') in whole_nm: #정상 데이터만 이렇게
      with open(os.path.join(annot_dirs[3], lbl_file),'r') as f:
        data = json.load(f)
        ifn = data["description"]["image"]
        ifn = ifn.replace('JPG','jpg') #대문자로 돼 있고 지랄
        if ifn in all_imgs[3]:
            #image_filename = data["description"]["image"]
            bbox = data["annotations"]["bbox"][0]
            bbox_x = int(bbox["x"])
            bbox_y = int(bbox["y"])
            bbox_w = int(bbox["w"])
            bbox_h = int(bbox["h"])
            bbox_x2 = bbox_x + bbox_w
            bbox_y2 = bbox_y + bbox_h
            class_name = f"{data['annotations']['disease']}_{crop_names[data['annotations']['crop']]}"
            class_name = class_name.replace('고추','pepper')
            value = [{
                  "x":bbox_x,
                  "y":bbox_y,
                  "x2":bbox_x2,
                  "y2":bbox_y2,
                  "id":str(id),
                  "label":class_name
            }]
            id += 1
            IBM["annotations"][ifn] = value

output_directory = "/content/drive/MyDrive/Capstone/AIHUB-dataset/104.식물 병 유발 통합 데이터/01.데이터/2.Validation/라벨링데이터/VL1_고추"
output_file_path = os.path.join(output_directory,'papper_ibm_annotations_n' + ".json")
with open(output_file_path,"w") as output_json_file:
  json.dump(IBM, output_json_file, ensure_ascii=False)
```
