```python
records_b2 = []

for page in project.search_all(
    class_name = "b2_tomato",
    in_dataset = True,
    fields = ["name"],
):
    records_b2.extend(page)

print(len(records_b2))
```

```python
fns_b2 = []
for r in records_b2:
  fns_b2.append(r["name"])
print(len(fns_b2))
```

```python
import random
random.shuffle(fns_b2)
```

```python
fns_b2 = fns_b2[:300]
```

```python
categories_list = [
    'b2_tomato'
]
```

```python
label_path = '/content/drive/MyDrive/Capstone/AIHUB-dataset/104.식물 병 유발 통합 데이터/01.데이터/2.Validation/라벨링데이터/VL5_토마토/생리장해'
```

```python
import os
import json
idx = 0
IBM = {
      "version": "1.0",
      "type": "localization",
      "labels": categories_list,
      "annotations":{}
}
for img in fns_b2:
  json_fn = img.replace('jpg','').replace('JPG','')
  json_fn += 'json'
  json_file_path = os.path.join(label_path,json_fn)
  with open(json_file_path,"r") as json_file:
      data = json.load(json_file)
      image_filename = data["description"]["image"]
      class_name = f"{data['annotations']['disease']}_tomato"
      parts = data["annotations"]["part"]
      value = []
      for p in parts:
        temp = {}
        temp["x"] = p["x"]
        temp["y"] = p["y"]
        temp["x2"] = p["x"]+p["w"]
        temp["y2"] = p["y"]+p["h"]
        temp["id"] = str(idx)
        idx += 1
        temp["label"] = class_name
        value.append(temp)
      IBM["annotations"][img] = value

with open('/content/drive/MyDrive/Capstone/AIHUB-dataset/104.식물 병 유발 통합 데이터/01.데이터/2.Validation/라벨링데이터/VL5_토마토/tomato-part-b2',"w") as output_json_file:
  json.dump(IBM,output_json_file, ensure_ascii=False)
```
