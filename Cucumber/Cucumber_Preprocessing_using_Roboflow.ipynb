```python
IBM = {
   "version": "1.0",
    "type": "localization",
    "labels": categories_list,
    "annotations":{}
}
id = 0
for x,annots_x in enumerate(all_annots):
  for lbl_file in annots_x:
    if lbl_file.endswith(".json"):
      with open(os.path.join(annot_dirs[x], lbl_file),'r') as f:
        data = json.load(f)
        ifn = data["description"]["image"]
        if ifn in all_imgs[x]:
          image_filename = data["description"]["image"]
          bbox = data["annotations"]["bbox"][0]
          bbox_x = int(bbox["x"])
          bbox_y = int(bbox["y"])
          bbox_w = int(bbox["w"])
          bbox_h = int(bbox["h"])
          bbox_x2 = bbox_x + bbox_w
          bbox_y2 = bbox_y + bbox_h
          class_name = f"{data['annotations']['disease']}_{crop_names[data['annotations']['crop']]}"
          class_name = class_name.replace('오이','cucumber')
          value = [{
                "x":bbox_x,
                "y":bbox_y,
                "x2":bbox_x2,
                "y2":bbox_y2,
                "id":str(id),
                "label":class_name
          }]
          id += 1
          IBM["annotations"][image_filename] = value

output_directory = "/content/drive/MyDrive/Capstone/AIHUB-dataset/104.식물 병 유발 통합 데이터/01.데이터/2.Validation/라벨링데이터/VL4_오이"
output_file_path = os.path.join(output_directory,'cucumber_diseases_annotations' + ".json")
with open(output_file_path,"w") as output_json_file:
  json.dump(IBM, output_json_file, ensure_ascii=False)
```
