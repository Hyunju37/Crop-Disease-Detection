```python
#처음 업로드
IBM = {
    "version": "1.0",
    "type": "localization",
    "labels": categories_list,
    "annotations":{}
}

nums = [2500, 2500, 1300, 2500]
idx = 0

for label_idx, label_path in enumerate(all_labels_path):
  this_labels = os.listdir(label_path) # 해당 클래스의 모든 json 파일들명
  this_imgs = os.listdir(all_imgs_path[label_idx]) # 해당 클래스의 모든 원천이미지 파일들명
  total = len(this_labels) #json 파일들 개수
  indices = list(range(total))
  indices = indices[:nums[label_idx]]
  random.shuffle(indices)
  for i in indices:
    json_file = this_labels[i]
    if json_file.endswith(".json"):
      input_file_path = os.path.join(label_path, json_file)

      with open(input_file_path,"r") as json_file:
        data = json.load(json_file)
      if data["description"]["image"] in this_imgs:
        image_filename = data["description"]["image"]
        upload_images[label_idx].append(image_filename)

        bbox = data["annotations"]["bbox"][0]
        bbox_x = int(bbox["x"])
        bbox_y = int(bbox["y"])
        bbox_w = int(bbox["w"])
        bbox_h = int(bbox["h"])
        bbox_x2 = bbox_x + bbox_w
        bbox_y2 = bbox_y + bbox_h
        class_name = f"{data['annotations']['disease']}_{crop_names[data['annotations']['crop']]}_r{data['annotations']['risk']}"
        class_name = class_name.replace('딸기','strawberry')

        value = [{
            "x":bbox_x,
            "y":bbox_y,
            "x2":bbox_x2,
            "y2":bbox_y2,
            "id":str(idx),
            "label":class_name
        }]
        idx += 1
        IBM["annotations"][image_filename] = value

output_file_path = os.path.join(output_directory,'strawberry_annotations' + ".json")
with open(output_file_path,"w") as output_json_file:
  json.dump(IBM, output_json_file, ensure_ascii=False)
```
