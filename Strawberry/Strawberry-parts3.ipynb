```python
from roboflow import Roboflow
rf = Roboflow(api_key="2nxKEUjSzlIMWDl3hWnV")
```

```python
project = rf.project("strawberry-rbo1c")
records_c1 = []
records_c2 = []

for page in project.search_all(
    class_name = "c1_strawberry",
    in_dataset = True,
    fields = ["name"],
):
    records_c1.extend(page)

for page in project.search_all(
    class_name = "c2_strawberry",
    in_dataset = True,
    fields = ["name"],
):
    records_c2.extend(page)

print(len(records_c1),len(records_c2))
```

```python
fns_c1 = []
for r in records_c1:
  fns_c1.append(r["name"])
fns_c2 = []
for r in records_c2:
  fns_c2.append(r["name"])
```

```python
print(len(fns_c1),len(fns_c2))
```

```python
project = rf.workspace().project("strawberry-temp")
```

```python
import random
random.shuffle(fns_c1)
boundary = int(len(fns_c1) * 4/5)
train_c1 = fns_c1[:boundary]
valid_c1 = fns_c1[boundary:]
print(len(train_c1),len(valid_c1))

random.shuffle(fns_c2)
boundary = int(len(fns_c2) * 4/5)
train_c2 = fns_c2[:boundary]
valid_c2 = fns_c2[boundary:]
print(len(train_c2),len(valid_c2))
```

```python
train_c1_sample = train_c1[:int(len(train_c1)*0.1)]
valid_c1_sample = valid_c1[:int(len(valid_c1)*0.1)]
train_c2_sample = train_c2[:int(len(train_c2)*0.1)]
valid_c2_sample = valid_c2[:int(len(valid_c2)*0.1)]
```

```python
print(len(train_c1_sample),len(train_c2_sample),len(valid_c1_sample),len(valid_c2_sample))
```
