#어노테이션파일만들기

```python
label_path = "/content/drive/MyDrive/Capstone/AIHUB-dataset/104.식물 병 유발 통합 데이터/01.데이터/2.Validation/라벨링데이터/VL4_오이/병해"
```

```python
categories_list = [
 'a4_cucumber'
]
```

```python
import os
import json
idx= 0
IBM = {
      "version": "1.0",
      "type": "localization",
      "labels": categories_list,
      "annotations":{}
}
for img in fns_a4_u:
  json_fn = img.replace('jpg','').replace('JPG','')
  json_fn += 'json'
  json_file_path = os.path.join(label_path,json_fn)
  with open(json_file_path,"r") as json_file:
      data = json.load(json_file)
      image_filename = data["description"]["image"]
      class_name = f"{data['annotations']['disease']}_cucumber"
      parts = data["annotations"]["part"]
      value = []
      for p in parts:
        temp = {}
        temp["x"] = p["x"]
        temp["y"] = p["y"]
        temp["x2"] = p["x"]+p["w"]
        temp["y2"] = p["y"]+p["h"]
        temp["id"] = str(idx)
        idx += 1
        temp["label"] = class_name
        value.append(temp)
      IBM["annotations"][img] = value

with open('/content/drive/MyDrive/Capstone/AIHUB-dataset/104.식물 병 유발 통합 데이터/01.데이터/2.Validation/라벨링데이터/VL4_오이/cucumber-part-a4.json',"w") as output_json_file:
  json.dump(IBM,output_json_file, ensure_ascii=False)
```
