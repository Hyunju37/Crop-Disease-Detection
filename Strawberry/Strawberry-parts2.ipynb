```python
project = rf.project("strawberry-rbo1c")
records_b = []
for page in project.search_all(
    class_name = "b1_strawberry",
    in_dataset = True,
    fields = ["name"],
):
    records_b.extend(page)
for page in project.search_all(
    class_name = "b6_strawberry",
    in_dataset = True,
    fields = ["name"],
):
    records_b.extend(page)
for page in project.search_all(
    class_name = "b7_strawberry",
    in_dataset = True,
    fields = ["name"],
):
    records_b.extend(page)
for page in project.search_all(
    class_name = "b8_strawberry",
    in_dataset = True,
    fields = ["name"],
):
    records_b.extend(page)

print(len(records_b))
```

```python
fns_b = []
for r in records_b:
  fns_b.append(r["name"])
```

```python
import random
random.shuffle(fns_b)
boundary = int(len(fns_b) * 4/5)
train = fns_b[:boundary]
test = fns_b[boundary:]
print(len(train),len(test))
```
